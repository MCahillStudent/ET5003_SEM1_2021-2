# Problem 7
Plot the graph of the function $\cos(8 x)$ in red and $x^2$ in blue for $x \in[0,1]$ and add the label "x" and "f(x)" to the respective columns using `fontsize` 14.

```python
x = np.arange(0, 1, 0.01)
y_1 = np.cos(8 * x)
y_2 = x * x

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.scatter(x, y_1, color='r')
ax.scatter(x, y_2, color='b')
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('f(x)',fontsize=14)
```

# Problem 8
Plot the graph of the function $\cos(8 x)$ in red and $x^2$ in blue for $x \in[0,1]$ and add a legend
with labels "cos(8x)" and "$x^2$"

Reference:
https://www.geeksforgeeks.org/matplotlib-pyplot-legend-in-python/

```python
x = np.arange(0, 1, 0.01)
y_1 = np.cos(8 * x)
y_2 = x * x

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.scatter(x, y_1, color='r', label ='cos(8x)')
ax.scatter(x, y_2, color='b', label = 'x$^2$')

# Add a legend  
plt.legend()

# Set label
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
```

# Problem 10
Plot the graph of the 2D function $e^{-x^2-y^2}$  for $x,y \in[-3,3]$.

Reference:
https://glowingpython.blogspot.com/2012/01/how-to-plot-two-variable-functions-with.html

```python
x = y = np.arange(-3, 3, 0.01)
```

```python
def z_func(x,y):
 return np.exp(-x**2 - y**2)

X, Y = np.meshgrid(x, y) # grid of point
Z = z_func(X, Y) # evaluation of the function on the grid

fig, ax = plt.subplots()
cnt = ax.contour(Z, cmap=plt.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1])
```
